```python
plt.subplot()
plt.plot(db_json["calc"], db_json["expt"], 'bo')
plt.title("Mobley : clac vs expt")
plt.xlabel("calc")
plt.ylabel("expt")
plt.legend(['Coefficient correlation = {:.3f}'.format(db_json["calc"].corr(db_json["expt"]))])
plt.plot(x, fit[0] * x + fit[1], color='red')
```
